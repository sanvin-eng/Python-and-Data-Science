```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
Data = pd.read_csv('CardioGoodFitness.csv')  # read the data
Data.head() 
```

```python
Data.shape #Gives usa basic shape of the data
```

# Data Preprocessing Using Pandas Profiling
Pandas Profiling is carried out as it helps in giving a basic analysis on the number of missing values in the data frame and in addition givees a Univariative and a Multivariative analysis.
As in this case, there's no need to clean up the data as there are no missing values.

```python
import pandas_profiling 
pandas_profiling.ProfileReport(Data)
```

```python
#In this step all the variables (Columns) from the data frame is copied to local variables for easier access
Product= Data['Product']
Age= Data['Age']
Gender= Data['Gender']
Education= Data['Education']
MaritalStatus= Data['MaritalStatus']
Usage= Data['Usage']
Fitness= Data['Fitness']
Income= Data['Income']
Miles= Data['Miles']
```

# Univariative Analysis

All Columns or features with a quantitative nature that is Product, Gender and MaritalStatus are analysed using bar graphs 

1. Product

```python
sns.countplot(Product) 
```

Observations:
    1. TM195 is the most bought by the customers followed by TM498 and TM798 respectively. 

2. Age 

This particular feature is analysed through a violin chart as this kind of graph helps in highlighting the age groups that widely comprises our Customer Base.

```python
sns.violinplot(Age) 
```

Observations:
    1. There's a high demand inetween ages 20-30.
    2. The gradually reduces as we go to older ages.

3. Gender

```python
sns.countplot(Gender)
```

Well to conclude, Males use more Cardio Good Fitness Treadmills than females.

4. Education

```python
sns.distplot(Education, hist=False, rug=True) 
```

Observations:
    1. There are sudden peaks at ages 14,16 and 18.
    2. Relatively sudden drops at ages 15, 17 and 19.
    3. Even in this case the older faction don't use Cardio Good Fitness Treadmills.

5. Mariatal Status

```python
sns.countplot(MaritalStatus)
```

Only a single conclusion overall partnered people use Cardio Good Fitness.

6. Usage

```python
sns.boxplot(Usage)
```

Observations: 
    1. Most usage occurs for around 3-4 years.
    2. Only 2 outliers in this scenario so in best and rare cases the machines do run for 6-7 years.

7. Fitness

```python
sns.countplot(Fitness)
```

Majority of the users using Cardio Good Fitness have a Fitness level of 3 which is a positive sign that tell's us that the machine are keeping majority of it's customers healthy.

8. Income

```python
sns.distplot(Income, hist=True, rug=True) 
```

Observations:
    1. Most of this data set is around the 50,000 annual income range.
    2. Cardio Good Fitness has most of it's customers around the 35,000 to 70,000 range.    

9. Miles 

```python
plt.figure(figsize=(20,10)) # makes the plot wider
plt.hist(Miles, color='y') # plots a simple histogram
plt.axvline(Miles.mean(), color='m', linewidth=1)
plt.axvline(Miles.median(), color='b', linestyle='dashed', linewidth=1)
plt.axvline(Miles.mode()[0], color='w', linestyle='dashed', linewidth=1)
```

Observations:
    1. The mean for the miles run by customers is a little north 100 miles.
    2. Most customers end up running 120 miles although highest is arounf 350.
    3. Median is in the 90's
    4. Mode is 80.
